```python
#Data Cleaning & Visualization
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
df=pd.read_csv("Personal Project Data.csv")
print(df)
```

```python
df.head() #Checking first 5 values of data 
```

```python
df.isnull()#Checking null values in data 
```

```python
df.duplicated() # Checking duplicate values in data
```

```python
df.drop_duplicates(inplace=True) #Droping duplicate values
print(df)
```

```python
df.sort_values(by='Quantity Sold',ascending=True) # Sorting data by Quantity Sold
```

```python
df.sort_values(by=['Date','Region'],ascending=True) # Sorting data by Date and Region
```

```python
#A time series line chart of Sales Amount over time.
sns.lineplot(data=df,x='Date',y='Sales Amount',marker='o',mec='red',mfc='yellow',hue='Product Category')
plt.xticks(rotation=47)
plt.grid(ls="--",color='black')
plt.ylabel('Sales Amount',fontsize=12)
plt.xlabel('Date',fontsize=12)
plt.title("Sales Over The Time",loc='right')
plt.annotate('Highest Sale',)
```

```python
#A bar chart comparing Sales Amount across different Product Categories.
sns.barplot(data=df,x='Product Category',y='Sales Amount',palette='vlag',hue='Region')
plt.grid(ls='--')
plt.title('Sales Amount of Product Categories',fontsize=13,loc='right')
plt.xlabel("Product Categories",fontsize=13)
plt.ylabel("Sales Amount",fontsize=13)
```

```python
#A scatter plot of Sales Amount versus Quantity Sold
sns.scatterplot(data=df,x='Sales Amount',y='Quantity Sold',color='red',hue='Product Category',palette='rocket')
plt.grid(ls="--")
plt.title("Sales Amount Vs Quantity Sold",fontsize=13,loc='right')
plt.xlabel("Sales Amount",fontsize=13)
plt.ylabel("Quantity Sold",fontsize=13)
```

```python
#Create a pie chart to show the percentage contribution of each Region to the total Sales Amount
region_sales = df.groupby('Region')['Sales Amount'].sum()
colors=('brown','darkgreen','maroon','orange')
plt.pie(region_sales,labels=region_sales.index,startangle=140,autopct='%1.1f%%',shadow=True,colors=colors)
plt.axis=('equal')
plt.title("Contribution of Each Region",fontsize=14,loc='right')
```
